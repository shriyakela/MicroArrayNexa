```python
!pip install scikit-image opencv-python

import os
import cv2
import numpy as np
from skimage.util import random_noise
from skimage import img_as_ubyte

from google.colab import files
files.upload()
!unzip dataset_images.zip


INPUT_DIR  = "/content/dataset_images"
OUTPUT_DIR = "/content/speckle-output-new"
import shutil
shutil.rmtree(OUTPUT_DIR, ignore_errors=True)
os.makedirs(OUTPUT_DIR, exist_ok=True)


SIGMAS = [0.05, 0.078, 0.106, 0.133, 0.161, 0.189]

ALLOWED_EXTS = {".png", ".jpg", ".jpeg", ".tif", ".tiff", ".bmp"}


os.makedirs(OUTPUT_DIR, exist_ok=True)

def read_and_normalize(path):
    img = cv2.imread(path, cv2.IMREAD_UNCHANGED)
    if img is None:
        raise RuntimeError(f"Cannot read: {path}")


    if img.ndim == 2:
        img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
    elif img.ndim == 3 and img.shape[2] == 1:
        img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)


    img = img.astype(np.float32)
    mn, mx = img.min(), img.max()
    if mx - mn > 0:
        img = (img - mn) / (mx - mn)
    else:
        img = np.zeros_like(img, dtype=np.float32)


    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img

def save_rgb_float_as_png(path, rgb_float):
    uint8 = img_as_ubyte(rgb_float)
    bgr = cv2.cvtColor(uint8, cv2.COLOR_RGB2BGR)
    cv2.imwrite(path, bgr)

files = sorted([f for f in os.listdir(INPUT_DIR) if os.path.splitext(f)[1].lower() in ALLOWED_EXTS])

if not files:
    print("No images found in", INPUT_DIR)
else:
    print(f"Found {len(files)} images. Generating {len(SIGMAS)} speckle variants each -> {len(files)*len(SIGMAS)} images.")

for fname in files:
    in_path = os.path.join(INPUT_DIR, fname)
    try:
        img = read_and_normalize(in_path)
    except Exception as e:
        print("Skipping", fname, ":", e)
        continue

    base = os.path.splitext(fname)[0]
    for i, sigma in enumerate(SIGMAS, start=1):

        noisy = random_noise(img, mode="speckle", var=(sigma ** 2))
        out_name = f"{base}_speckle_{i:02d}_sigma_{sigma:.3f}.png"
        out_path = os.path.join(OUTPUT_DIR, out_name)
        save_rgb_float_as_png(out_path, noisy)

print("Done.")
```
